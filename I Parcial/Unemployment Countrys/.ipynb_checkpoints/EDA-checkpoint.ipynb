```python
import pandas as pd
import numpy as np
```

```python
#Importacion de los dataframes
dfunemployment = pd.read_csv("unemployment analysis.csv", delimiter=",")
dfcountry = pd.read_csv("country information.csv", delimiter=",")
dfcontinents = pd.read_csv("continents.csv", delimiter=",")
```

```python
#Modificacion del dataframe unemployment
dfmodified = pd.DataFrame()
for y in dfunemployment.columns[ 2 :  ]:
    dftemp  = dfunemployment.loc[ :  , ['Country Name','Country Code', y] ]
    dftemp['Year'] = y
    dftemp.columns = ['Country Name', 'Country Code', 'Value', 'Year']
    dftemp = dftemp.loc[ : , ['Country Name', 'Country Code', 'Year', 'Value'] ]
    
    dfmodified = pd.concat([dfmodified, dftemp])
```

```python
#Merge entre los diferentes dataframes
df_union_1 = pd.merge(dfmodified,dfcountry[['country_alpha3_code','capital', 'country_demonym', 'total_area', 'population', 'currency_name', 'currency_symbol', 'lang_name']], left_on = "Country Code", right_on = "country_alpha3_code" , how = 'left')
df_union_2 = pd.merge(df_union_1, dfcontinents[['alpha-3', 'region', 'sub-region']], left_on = "Country Code", right_on = "alpha-3" , how = 'left')
```

```python
#Eliminando columnas que tienen los datos de las llaves que se utilizaron para el merge que repiten datos innecesariamente
dffinal = df_union_2.drop(columns = ['country_alpha3_code', 'alpha-3'])
```
