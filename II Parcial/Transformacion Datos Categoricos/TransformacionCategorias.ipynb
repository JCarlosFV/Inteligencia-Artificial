```python
import pandas as pd
```

```python
adidas = pd.read_csv("adidas.csv")
```

```python
adidas.columns
```

```python
cols = ['url', 'name', 'sku', 'currency',
       'availability', 'color', 'category', 'source', 'source_website',
       'breadcrumbs', 'description', 'brand', 'images', 'country', 'language']
```

```python
#Ejercicio 1

cols_validas = []
cols_many_levels = []
cols_one_level = []
cols_levels_very_representative = []

size_df = adidas.shape[0]
umbral = .95

for c_ in cols:
    niveles = adidas.loc[: , [c_] ].drop_duplicates().shape[0]
    
    if ( niveles / size_df ) >= umbral:
        cols_many_levels.append(c_)
    elif niveles == 1:
        cols_one_level.append(c_)
    else:
        niveles_group = adidas.groupby( [c_], as_index = False ).agg( {"sku" : ["count"] } )
        niveles_group.columns = niveles_group.columns.droplevel(1)
        niveles_group["Porcentaje"] = niveles_group["sku"] / adidas.shape[0]
        niveles_group = niveles_group.sort_values(["Porcentaje"] , ascending = False)
        
        if (niveles_group["Porcentaje"].iloc[0] >= 0.98):
            cols_levels_very_representative.append(c_)
        else:
            cols_validas.append(c_)
```

```python
#Ejercicio 2

columnas_nuevas = []

for c_ in cols:
    niveles = adidas.loc[: , [c_] ].drop_duplicates().shape[0]
    
    if (niveles >= 3) & (niveles <= 6):
        niveles_group = adidas.groupby( [c_], as_index = False ).agg( {"sku" : ["count"] } )
        niveles_group.columns = niveles_group.columns.droplevel(1)
        niveles_group["Porcentaje"] = niveles_group["sku"] / adidas.shape[0]
        niveles_group = niveles_group.sort_values(["Porcentaje"] , ascending = True)
        
        nivel_bajo = niveles_group["Porcentaje"].iloc[0]
        niveles_group[c_ + "_new"] = niveles_group.apply( lambda r: ("Others" if r["Porcentaje"] == nivel_bajo else r[c_] ) , axis = 1)
        niveles_group = niveles_group.drop( ["sku", "Porcentaje"], axis = 1 )
        columnas_nuevas.append(niveles_group.columns[1])
        
        adidas = adidas.merge(niveles_group, on = [c_], how = "left")
```
